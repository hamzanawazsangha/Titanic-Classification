```python
def my_best_model(x, y):
    algos = {
        'logistic_regression': {
            'model': LogisticRegression(),
            'params': {
                'fit_intercept': [True, False],
                'C': [0.1, 1, 10],
                'solver': ['lbfgs', 'liblinear']
            }
        },
        'Decision_Tree_Classifier': {
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion': ['gini', 'entropy'],
                'splitter': ['best', 'random'],
                'max_depth': [None, 10, 20, 30],
                'min_samples_split': [2, 10, 20]
            }
        },
        'GaussianNB': {
            'model': GaussianNB(),
            'params': {
                'priors': [None]
            }
        },
        'Random_Forest_Classifier': {
            'model': RandomForestClassifier(),
            'params': {
                'n_estimators': [ 200, 300],
                'max_depth': [10, 20],
                'min_samples_split': [2, 10]
            }
        },
        'Support_Vector_Classification': {
            'model': SVC(),
            'params': {
                'gamma': ['scale', 'auto'],
                'C': [0.1, 1, 10],
                'kernel': ['linear', 'rbf']
            }
        },
        'KNN': {
            'model': KNeighborsClassifier(),
            'params': {
                'n_neighbors': [3, 5, 7, 9],
                'weights': ['uniform', 'distance'],
                'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
df6=pd.DataFrame()
```

```python
df6 = my_best_model(x,y)
df6
```

```python
df6.best_params[3]
```

```python
def plot_best_scores(df):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='model', y='best_score', hue='model', data=df, palette='viridis', dodge=False)
    plt.xlabel('Model')
    plt.ylabel('Best Score')
    plt.title('Best Scores of Different Algorithms')
    plt.legend([],[], frameon=False)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_best_scores(df6)
```
